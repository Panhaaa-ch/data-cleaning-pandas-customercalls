```python
import pandas as pd
df = pd.read_excel('/Users/macbook/Downloads/Training_1/Customer Call List.xlsx')
df
```

```python
df = df.drop(columns=['Not_Useful_Column'])
df
```

```python
df = df.drop_duplicates()
df
```

```python
# Cleaning Last Name Column
# df['Last_Name'] = df['Last_Name'].str.strip()
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name']
df['Last_Name'] = df['Last_Name'].str.strip('. /')
df


```

```python
# Manipulating the Phone Number and appplying data formatting
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
# df['Phone_Number'] = df['Phone_Number'].str.replace('nan', '')
# df['Phone_Number'] = df['Phone_Number'].str.replace('Na', '')
# df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])
# df['Phone_Number'] = df['Phone_Number'].str.replace('--', '')
# df['Phone_Number']

# Tidy Method
df['Phone_Number'] = df['Phone_Number'].fillna('')
df['Phone_Number'] = (df['Phone_Number'].astype(str).str.replace('[^0-9]', '', regex=True))   # strip everything except digits in one go
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if len(x) == 10 else x)
df['Phone_Number'].unique()




```

```python
# Separate the Address
df[['Street', 'City', 'Postal_Code']] = df['Address'].str.split(',', n = 2, expand = True)
df = df.drop(columns=['Address'])
df
```

```python
# Replace Yes and No with Y and N correspondingly
df['Paying Customer'] = df['Paying Customer'].apply(lambda x: "Y" if x in ['Yes', 'Y'] 
                                                    else ("N" if x in['No', 'N'] else x))
df
```

```python
# Replace Yes and No with Y and N
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: 'Y' if x in ['Yes', 'Y'] 
                                                else('N' if x in ['No', 'N'] else x))
df
                                
```

```python
# Remove empty phone numbers
df = df[df['Phone_Number'] != '']
# Remove rows with Do not Contact Y
df = df[df['Do_Not_Contact'] == 'N']
# Reset index count

df = df.reset_index(drop=True)
df
```

```python
# Filling Nan values with meaningufl data
df['City'] = df['City'].fillna('Unknown')
# Drop the Postal Code because each row is almost empty
df = df.drop(columns=['Postal_Code'])
# Display First 10 values
df.head(10)
```
